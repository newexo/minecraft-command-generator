# Minecraft Command Generator - Give Commands Demo

This notebook demonstrates how to use the minecraft-command-generator library to create various give commands.

## Setup

First, let's import the necessary modules.

```python
from mc_commands.item import Item
from mc_commands.give import GiveCommand, GiveCount, GiveItem
from mc_commands.target import NEAREST_PLAYER, SELF, ALL_PLAYERS, RANDOM_PLAYER
from mc_commands.loaders import get_item, load_items
from mc_commands.stacking import break_into_stacks, shulker_give
```

## Creating Items Manually

You can create Item objects directly by specifying the item properties.

```python
# Create a diamond item
diamond = Item(id=264, name="diamond", display_name="Diamond", stack_size=64)
print(f"Item: {diamond.display_name}")
print(f"Name: {diamond.name}")
print(f"Max stack size: {diamond.stack_size}")
```

## Loading Items from Minecraft Data

You can load real item data from the minecraft-data package for a specific Minecraft version.

```python
# Load all items for Minecraft 1.21.6
all_items = load_items(version="1.21.6")
print(f"Total items available: {len(all_items)}")

# Get a specific item
diamond = get_item("diamond", version="1.21.6")
print(f"\nDiamond from data:")
print(f"  Display name: {diamond.display_name}")
print(f"  Stack size: {diamond.stack_size}")
```

## Creating Simple Give Commands

The simplest case: give a single item to a player.

```python
# Give 1 diamond to nearest player
cmd = GiveCommand(
    targets=NEAREST_PLAYER,
    item=GiveItem(item=diamond),
)
print(f"Command: {cmd}")
```

## Give Commands with Quantities

You can specify different quantities for items.

```python
# Give 32 diamonds to all players
cmd = GiveCommand(
    targets=ALL_PLAYERS,
    item=GiveItem(item=diamond),
    count=GiveCount(count=32),
)
print(f"Command: {cmd}")

# Give 5 diamonds to self
cmd = GiveCommand(
    targets=SELF,
    item=GiveItem(item=diamond),
    count=GiveCount(count=5),
)
print(f"Command: {cmd}")
```

## Give Commands with Item Components

You can add data components to items (like enchantments).

```python
# Give an enchanted diamond sword
sword = Item(id=268, name="diamond_sword", display_name="Diamond Sword", stack_size=1)

cmd = GiveCommand(
    targets=NEAREST_PLAYER,
    item=GiveItem(
        item=sword,
        components='[enchantments={"minecraft:sharpness":5}]',
    ),
)
print(f"Command: {cmd}")
```

## Breaking Items into Stacks

The `break_into_stacks` utility helps you break quantities into properly-sized stacks.

```python
# Break 129 diamonds into stacks (max 64 per stack)
stacks = break_into_stacks([(diamond, 129)])
print(f"Breaking 129 diamonds into stacks:")
for item, count in stacks:
    print(f"  - {count} {item.name}")

# Multiple items with different stack sizes
stick = get_item("stick", version="1.21.6")
egg = get_item("egg", version="1.21.6")

stacks = break_into_stacks([
    (diamond, 100),
    (stick, 200),
    (egg, 50),
])
print(f"\nBreaking multiple items:")
for item, count in stacks:
    print(f"  - {count} {item.name}")
```

## Creating Shulker Box Commands

Use `shulker_give` to create commands that give a shulker box filled with items.

```python
# Create a shulker box item
shulker_box = Item(id=454, name="shulker_box", display_name="Shulker Box", stack_size=1)

# Create a shulker box with 64 diamonds and 32 sticks
cmd = shulker_give(
    items=[(diamond, 64), (stick, 32)],
    shulker_box=shulker_box,
    target=NEAREST_PLAYER,
)
print(f"Shulker box command:\n{cmd}")
print(f"\nCommand length: {len(str(cmd))} characters")
```

## Complex Shulker Box Example

You can populate a shulker box with many items (up to 27 slots).

```python
# Load several items from minecraft-data
items_to_pack = [
    (get_item("diamond", "1.21.6"), 129),
    (get_item("stick", "1.21.6"), 200),
    (get_item("egg", "1.21.6"), 50),
    (get_item("iron_ingot", "1.21.6"), 64),
    (get_item("gold_ingot", "1.21.6"), 48),
]

# Create the shulker box command
cmd = shulker_give(
    items=items_to_pack,
    shulker_box=shulker_box,
    target=NEAREST_PLAYER,
)

# Show the command
command_str = str(cmd)
print(f"Generated command:")
print(f"{command_str}")
print(f"\nCommand stats:")
print(f"  Length: {len(command_str)} characters")
print(f"  Total stacks: {len(cmd.item.components.split('{slot:'))}  # rough estimate")
```

## Copy Commands to Clipboard

You can easily copy generated commands for use in Minecraft.

```python
import pyperclip

# Generate a command
cmd = GiveCommand(
    targets=NEAREST_PLAYER,
    item=GiveItem(
        item=get_item("diamond", "1.21.6"),
    ),
    count=GiveCount(count=64),
)

# Copy to clipboard (requires pyperclip: pip install pyperclip)
try:
    pyperclip.copy(str(cmd))
    print(f"✓ Copied to clipboard: {cmd}")
except:
    print(f"Command: {cmd}")
    print("(Install pyperclip to copy commands to clipboard)")
```
